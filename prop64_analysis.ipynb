# Charts and Graphs

```python
#supporters of the proposition
top_supporters = support64.groupby(
    ["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters.head()
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

# Cleaning

```python
# cleaning multiple entries
def combine_names(row):
    if row.contributor_fullname.startswith('SEAN PARKER'):
        return 'SEAN PARKER'
    return row.contributor_fullname
```

```python
top_supporters['contributor_cleanname'] = top_supporters.apply(combine_names, axis='columns')
#top_supporters.apply 'applies' the combine_name function to all elements. axis=1 or columns means apply to each row
```

```python
top_supporters.head()
```

```python
ts=top_supporters.groupby(
    "contributor_cleanname"
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
chart = ts.amount.plot.barh()
chart.set_yticklabels(ts.contributor_cleanname)
```
